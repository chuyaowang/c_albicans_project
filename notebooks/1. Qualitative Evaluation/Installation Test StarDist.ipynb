```python
# GPU warm up
import numpy as np

# Load a pretrained StarDist model (example: versatile fluorescent nuclei)
model = StarDist2D.from_pretrained('2D_versatile_fluo')

# Create a dummy input image (e.g., 512x512 grayscale)
dummy_img = np.random.rand(512, 512).astype(np.float32)

# Warm-up run: one forward pass (not used for results)
_ = model.predict_instances(dummy_img)

print("GPU warm-up complete. Ready for real data!")
```

```python
from stardist.data import test_image_nuclei_2d
from stardist.plot import render_label
from csbdeep.utils import normalize
import matplotlib.pyplot as plt

img = test_image_nuclei_2d()

labels, _ = model.predict_instances(normalize(img))

plt.subplot(1,2,1)
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.title("input image")

plt.subplot(1,2,2)
plt.imshow(render_label(labels, img=img))
plt.axis("off")
plt.title("prediction + input overlay")
```
